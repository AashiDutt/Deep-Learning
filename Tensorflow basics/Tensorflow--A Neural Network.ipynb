# Example Neural Network

```python
n_features = 10
n_dense_neurons = 3  # only one layer
```

```python
x = tf.placeholder(tf.float32,(None,n_features))
```

```python
W = tf.Variable(tf.random_normal([n_features,n_dense_neurons]))   # Weights

b = tf.Variable(tf.ones([n_dense_neurons]))

```

```python
xW = tf.matmul(x,W) # xW
```

```python
z =tf.add(xW,b)     # xW + b
```

```python
a = tf.sigmoid(z)   
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    layer_out = sess.run(a,feed_dict={x : np.random.random([1,n_features])})
```

```python
print(layer_out)
```

## Full Network Example

Let's work on a regression example, we are trying to solve a very simple equation:

y = mx + b

y will be the y_labels and x is the x_data. We are trying to figure out the slope and the intercept for the line that best fits our data!

```python
# Simple Regression Problem
```

```python
x_data =np.linspace(0,10,10) + np.random_uniform(-1.5,1.5,10)

#np.random_uniform(-1.5,1.5,10) is noise
```

```python
x_data
```

```python
y_label = np.linspace(0,10,10) + np.random.uniform(-1.5,1.5,10)
```

```python
y_label
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(x_data, y_label,'*')
```

```python
np.random.rand(2)
```

```python
# y = mx + b

m = tf.Variable(0.68)

b = tf.Variable(0.52)
```

```python
error = 0

for x ,y in zip(x_data,y_label):
    
    y_hat = m*x + b
    
    error+= (y-y_hat)**2  # minimizing error
    
    
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.001)
train = optimizer.minimize(error)
```

```python
# global initializer for variables

init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    
    sess.run(init)
    
    epochs = 100  # epochs = no. of training steps
    
    for i in range(epochs):
        
        sess.run(train)
        

    # Fetch Back Results
    final_slope , final_intercept = sess.run([m,b])
    
```

```python
final_slope
```

```python
final_intercept
```

```python
x_test = np.linspace(-1,11,10)

# y = m*x + b
y_pred_plot = final_slope*x_test + final_intercept

plt.plot(x_test,y_pred_plot,'r')

plt.plot(x_data,y_label,'*')
```
