```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import keras.backend as K
from keras.callbacks import EarlyStopping
```

```python
from keras.datasets import imdb
```

```python
(X_train, y_train), (X_test, y_test) = imdb.load_data('imdb.npz',
                                                      num_words=None,
                                                      skip_top=0,
                                                      maxlen=None,
                                                      start_char=1,
                                                      oov_char=2,
                                                      index_from=3) # we start the index from 3
```

```python
# creating reverese index where we swap the keys and the values, we create a new dictionary

rev_idx = {v+3:k for k,v in idx.items()} 

# we started index from 3 so we do not miss on any value of index
```

```python
rev_idx

# index reversed
```

```python
# adding the left over characters

rev_idx[0] = 'padding_char'
rev_idx[1] = 'start_char'
rev_idx[2] = 'oov_char'
rev_idx[3] = 'unk_char'
```

```python
# all the reviews have diffrent lengths so we scale them to equal lengths.
from keras.preprocessing.sequence import pad_sequences
from keras.layers import LSTM
```

```python
maxlen = 100

X_train_pad = pad_sequences(X_train,maxlen=maxlen)
X_test_pad = pad_sequences(X_test, maxlen = maxlen)
```

```python
X_train_pad.shape
```

```python
X_train_pad[0]   # first padded review
```

```python
X_train[0]   # orignal review (without padding)
```

```python
max_features= max([max(x) for x in X_train_pad] + [max(x) for x in X_test_pad]) + 1
max_features

# maximum no. of elements present in the dictonary are:
```

```python
y_train  # a long sequence of ones and zeros
```

```python
from keras.layers import Embedding
```

```python
# building the model

K.clear_session()

model= Sequential()

model.add(Embedding(max_features ,128))   #we will embedd  our 88587 words into 128 vectors.

model.add(LSTM(64, dropout =0.2, recurrent_dropout=0.2 ))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# training model

model.fit(X_train_pad,y_train,
         batch_size=32,
         epochs=2,
         validation_split = 0.3)
```

```python
score, acc = model.evaluate(X_test_pad, y_test)
print('Test score:', score)
print('Test accuracy:', acc)
```
